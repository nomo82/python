```python
#Traemos el archivo
df_poblacion=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2025/_Python/population_total.csv', encoding='ISO-8859-1')
df_poblacion.head(3)
```

```python
#lo primero que vamos a hacer es eliminar los null o los vacios, con el metodod dropna
df_poblacion.dropna()
```

```python
#sobre escribo el dataframe
df_poblacion.dropna(inplace=True)
```

```python
df_poblacion
```

```python
#primero pivoteamos para que la visulizacion sea mas amigable
#recordemos que para pivotear tenemos 3 argumentos, el index, las columnas y valores
df_poblacion.pivot(index="year", columns="country", values="population")
```

```python
#creo un nuevo dataframe con la tabla que pivotee
df_poblacion_pivote=df_poblacion.pivot(index="year", columns="country", values="population")
```

```python
df_poblacion_pivote
```

```python
#ahora vamos a llamar unos paises para notrabajar con todo el dataframe yblo sobreescribo
df_poblacion_pivote=df_poblacion_pivote[['Colombia', 'United States', 'India', 'China', 'Indonesia']]
df_poblacion_pivote
```

```python
#vamos a hacer una grafica de lineas
#usaremos el argumeto kind
#el metodo para graficar es plot
#podemos personalizar la grafica con otros argumentos
#xlabel me sirve para darle normble al eje x, y laber para el y
#title sirve para asignar titulo a la grafica
#figzise sirve para modificar eltamaño del lienzo, hay muchos mas
df_poblacion_pivote.plot(kind='line', xlabel='Año', ylabel='Población',
                        title='Población 1955-2020', figsize=(10,5))
```
