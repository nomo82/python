Para poder usar Dataframes vamos a activar dos bibliotecas, una de ellas es Nompy y la otra es Pandas, las dos se complementan y hacen cosas diferentes. Para crear Dataframe existen varias posibilidades sin embargo la mas usada es la lectura directa desde una archivo


```python
import pandas as pd #Importo pandas y le pongo el sobre sobrenombre pd
import numpy as np #Importo Numpy y le pongo el sobrenombre np
```

```python
#La primera forma de crar un dataframe es usando el método array, luego guardarlo en una variable
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#Para crear el dataframe del arreglo anterior debemos usar Pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2=np.array([[3,6,9],[12,15,18],[21,24,27]])
```

```python
pd.DataFrame(dataframe_2)
```

```python
#Ahora vamos a cambiarle el nombre a las columnas, esto se hace de la siguiente manera
df2=pd.DataFrame(dataframe_2, index =['fila1','fila2','fila3'],
                 columns=['columna1','columna2','columna3'])
```

```python
df2
```
