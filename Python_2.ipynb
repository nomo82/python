Para poder usar Dataframe vamos a activar dos bibliotecas una de ellas en numpy y la otra es pandas, las dos se complementa, y hacen las cosas diferentes, para crear dataframe existen varias posibilidades sin embargo la mas usada es la lectura directa desde un archivo

```python
#Importo panda y le pongo el sobrenombre pd para no tener que escribir el nombre completo
import pandas as pd
import numpy as np
```

```python
#la primera forma de crear un dataframe es usando el metodo array y luego guardarlo en un avariable
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#para crear el dataframe del arreglo anterior debemos usar pandas
pd.DataFrame(dataframe_1)
```

```python
dataframe_2= np.array([[5,10,15],[20,25,30],[35,40,45],[50,55,60]])
pd.DataFrame(dataframe_2)
```

```python
dataframe_3= np.array([[3,6,9],[12,15,18],[21,24,27],[30,33,36]])
pd.DataFrame(dataframe_3)
```

```python
#Ahora vamos a cambiarle el nombre a las columnas, esto se hace de la siguiente manera,
df1=pd.DataFrame(dataframe_1, index=["fila1","fila2","fila3","fila4"],
                 columns=["columna1","columna2"])
df1
```

```python
df2=pd.DataFrame(dataframe_2, index=["fila1","fila2","fila3","fila4"],
                 columns=["columna1","columna2","columna3"])
df2
```

```python
df3=pd.DataFrame(dataframe_3, index=["fila1","fila2","fila3","fila4"],
                 columns=["columna1","columna2","columna3"])
df3
```

```python
#Primero creamos dos listas, puedo crear cuantas listas quiera entendiendo que las listas son las columnas del datafrane
ciudadColombia=["Bogota","Medellin","Cali","Barranquilla"]
CiudadesInfo=[11500000,2500000,2200000,1300000]
```
