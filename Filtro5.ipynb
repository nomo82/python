```python
#Usando isin vamos a filtar por varias condicione. En este caso vamos a realizarlos por NoteBooks o UltraBooks y Aplle o Asus
#Primero creamos las condiciones
filtro_1=df_computadores['Company'].isin(['Apple','Asus'])
filtro_2=df_computadores['TypeName'].isin(['Notebook','Ultrabook>'])
```

```python
#Ahora si filtramos los datos
df_computadores[filtro_1 & filtro_2]
```

```python
#Ahora contamos el resultado del DF
df_computadores[filtro_2 & filtro_2].value_counts('TypeName')
```
