```python
# Instalar bibliotecas necesarias
!pip install gspread oauth2client pandas --quiet

# Importar bibliotecas
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
import gspread
from google.colab import auth
from google.colab import files
from datetime import datetime

# Autenticación de Google Colab
auth.authenticate_user()

# Subir archivo JSON de credenciales
uploaded = files.upload()

# Configuración de acceso
scope = [
    'https://www.googleapis.com/auth/spreadsheets',
    'https://www.googleapis.com/auth/drive'
]
credentials = ServiceAccountCredentials.from_json_keyfile_name(list(uploaded.keys())[0], scope)
gc = gspread.authorize(credentials)

# ID del archivo de Google Sheets (extraído de la URL)
spreadsheet_id = '1VV3N462uxqijYpSH5bZFwBqh2VBsChE0KCtogfs1Nno'

# Abrir el archivo de Google Sheets
try:
    spreadsheet = gc.open_by_key(spreadsheet_id)
except Exception as e:
    print(f"Error al abrir el archivo: {e}")
    raise

# Crear una lista para almacenar los datos combinados
data_frames = []

# Función para convertir fechas al formato dd/mm/aaaa
def convertir_formato_fecha(valor):
    try:
        return datetime.strptime(valor, '%Y-%m-%d').strftime('%d/%m/%Y')
    except ValueError:
        try:
            return datetime.strptime(valor, '%d/%m/%Y').strftime('%d/%m/%Y')
        except ValueError:
            return valor  # Retorna el valor original si no es una fecha válida

# Recorrer todas las pestañas del archivo
for worksheet in spreadsheet.worksheets():
    sheet_name = worksheet.title
    print(f"Procesando pestaña: {sheet_name}")
    try:
        # Obtener los datos de la hoja
        data = worksheet.get_all_values()

        if data:  # Verificar si hay datos en la hoja
            headers = data[0]  # Encabezados
            content = data[1:]  # Filas sin encabezados

            if headers:  # Validar que los encabezados no estén vacíos
                df = pd.DataFrame(content, columns=headers)

                # Normalizar las columnas relevantes
                df['Nombre'] = df.iloc[:, 1]  # Columna B (Nombres)
                df['Teléfono'] = df.iloc[:, 2]  # Columna C (Teléfonos)
                df['Correo Electrónico'] = df.iloc[:, 3]  # Columna D (Correos)
                df['Fecha'] = df.iloc[:, 0]  # Columna A (Fechas)

                # Convertir la columna de fechas al formato dd/mm/aaaa
                df['Fecha'] = df['Fecha'].apply(convertir_formato_fecha)

                # Filtrar solo las columnas requeridas y agregar metadatos
                df = df[['Fecha', 'Nombre', 'Teléfono', 'Correo Electrónico']]
                df['Nombre de pestaña'] = sheet_name

                # Agregar al listado de DataFrames
                data_frames.append(df)
        else:
            print(f"La pestaña '{sheet_name}' está vacía.")
    except Exception as e:
        print(f"Error procesando la pestaña '{sheet_name}': {e}")
        continue

# Concatenar todos los DataFrames y exportar
if data_frames:
    all_data = pd.concat(data_frames, ignore_index=True)
    try:
        all_data.to_csv('datos_combinados_google_ads.csv', index=False, sep=';', encoding='utf-8')
        print('Datos exportados a datos_combinados_google_ads.csv con separador ";" y codificación UTF-8')
    except Exception as e:
        print(f"Error al exportar datos: {e}")
else:
    print("No se encontraron datos para exportar.")
```
