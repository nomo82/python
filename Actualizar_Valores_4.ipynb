```python
#Traemos el archivo DF Jugadores
#Leer el archivo CSV palyers 20.csv
df_jugadores=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2025/_Python/players_20.csv')
df_jugadores.head(5)
```

```python
#Ahora fijamo el index
df_jugadores=df_jugadores.set_index('short_name')
```

```python
#Traemos las columnas que vamos
df_jugadores=df_jugadores[['player_url', 'long_name',	'age', 'dob',	'height_cm',	'weight_kg', 'nationality',	'club']]
```

```python
df_jugadores.head(10)
```

```python
#axis nos sirve para especificar si quiero realizar la accion sobre la columna o la fila. Si es 0 sobre la fila y si es 1 sobre la columna
df_jugadores.drop('L. Messi', axis=0)
```

```python
#Vamos a dropear "Eliminar" teniendo en cuienta el index
df_jugadores.drop(index=['Neymar Jr']) #Como es el index se entiende que debe eliminar toida la fila de Neimar Jr
```

```python
#Ahora vamos a eliminar dos o mas filas y las vamos a actualizar y gardar en el DF
df_jugadores=df_jugadores.drop(['L. Messi', 'Neymar Jr'], axis=0)
```

```python
df_jugadores
```

```python
#Ahora vamos a eliminar una columna completa
df_jugadores = df_jugadores.drop('players_url', axis=1) #Cuando axis es igual a uno, quiere decir que eliminar toda la columna
```

```python
#para eliminar la última columna por posición y no lo supieramos los hacemos de la sigueitne manera
df_jugadores = df_jugadores.drop(df_jugadores.columns[-1], axis=1)
df_jugadores
```
