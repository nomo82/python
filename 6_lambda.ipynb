```python
#Traemos el mismo archivo
df_jugadores=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2025/_Python/players_20.csv')
#Le defino el index
df_jugadores.set_index('short_name', inplace=True)
#Selecciono las columnas que voy a usar
df_jugadores= df_jugadores[['player_url','long_name', 'age',  'dob',  'height_cm',  'weight_kg',  'nationality',  'club']]
df_jugadores.head(5)
```

```python
#Las funciones normalmente se ven de la siguiente manera
def sumar(a, b):
  x=a+b
  return (x)
```

```python
#Vamos a ver como se hace las funciones usando lambda
#Primero se asigna el nombre luego asignamos el valor, luego introducimos los parametros, luego dos puntos y por ultimo el output
sumar_2=lambda a,b: a+b
```

```python
sumar_2(39,39)
```

```python
modulo_2=lambda a,b: a%b
```

```python
modulo_2(48,2)
```

```python
#Ahora usemos a apply con lamba
#Vamosa a pasar los cm a mtros en la información de los jugadores
#En este caso si usamos apply no necesitamos poner el nombre de la función
df_jugadores['height_cm'].apply(lambda x:x/100)
```

```python
df_jugadores['age'].apply(lambda x:x*12)
```

```python
#Para convertir los cm a metros
df_jugadores['height_cm']/100
```

```python
#Tambien podemos pasar los nombre de los jugadores a mayusculas
df_jugadores['long_name'].apply(lambda x:x.upper())
```

```python
#Si quiero hacerlo simple pero con mas memoria sería asi
df_jugadores['long_name'].str.upper()
```

```python
#Tambien puedo usar lambda para extraer partes de fechas
#Primero debemos garantizar que la columna fecha este en modo date
df_jugadores['dob']=pd.to_datetime(df_jugadores['dob'])
```

```python
#ahora extraemos los años solo con la funcion lambda
df_jugadores['dob'].apply(lambda x:x.year)
```

```python
df_jugadores['dob'].apply(lambda x:x.month_name())
```

```python
df_jugadores['IMC'] = df_jugadores.apply(lambda x: x['weight_kg'] / ((x['height_cm'] / 100) ** 2), axis=1)
```
