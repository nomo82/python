Escribe un programa que calcule el total de uina compra de supermercado. El usuario debe ingresar el precio de dos productos y las cantidades que desee comprar de cada uno. Luego el programada debera permitir al usuario: Sumar los precios de cada productos. Multiplicar los precios por la cantidad de productos para obtener el costo total, restar un precio de descuento del precio total si el usuario tiene un cupon. Dividir el total entre dos personas si desean dividir la cuenta.

```python
#Funbciones Multiplicar, suma , resta y división asi se debe construir
def suma(a,b):
  resultado_suma=(a+b)
  return resultado_suma

def multiplicacion(a,b):
  resultado_multiplicacion=(a*b)
  return resultado_multiplicacion

def division(a,b):
  resultado_division=(a/b)
  return resultado_division

def resta(a,b):
  resultado_resta=(a-b)
  return resultado_resta
```

```python
suma(5,7)
```

```python
resta(5,7)
```

```python
multiplicacion(5,7)
```

```python
division(7,5)
```

```python
print('Bienvenido a Supermercados Michi')
precio_1 = int(input('Ingrese el precio del producto A: '))
cantidad_1 = int(input('Ingrese la cantidad del producto A: '))
precio_2 = int(input('Ingrese el precio del producto B: '))
cantidad_2 = int(input('Ingrese la cantidad del producto B: '))

#Opciones del programa
operator = input('''Elige una opción:
1. Suma de los precios de ambos productos
2. Calcular el costo total (precio por cantidad)
3. Aplicar el descuento al total
4. Dividir el total entre 2 personas
No olvides que solo puedes elegir 1, 2, 3 o 4: ''')

#Lógica según las opciones
if operator == '1':
    total_precio = suma(precio_1, precio_2)
    print(total_precio)

elif operator == '2':
    total_precio = suma(multiplicacion(precio_1, cantidad_1), multiplicacion(precio_2, cantidad_2))
    print(total_precio)

elif operator == '3':
    total_precio = suma(multiplicacion(precio_1, cantidad_1), multiplicacion(precio_2, cantidad_2))
    descuento = multiplicacion(total_precio, 0.10)
    total_descuento = resta(total_precio, descuento)
    print(f'El total con descuento es: {total_descuento}')

elif operator == '4':
    total_precio = suma(multiplicacion(precio_1, cantidad_1), multiplicacion(precio_2, cantidad_2))
    descuento = multiplicacion(total_precio, 0.10)
    total_descuento = resta(total_precio, descuento)
    total_persona = division(total_descuento, 2)
    print(f'Cada persona debe cancelar: {total_persona}')

else:
    print('Opción no válida, por favor selecciona 1, 2, 3 o 4.')
```
