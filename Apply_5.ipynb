```python
#Traemos el archivo DF Jugadores
#Leer el archivo CSV palyers 20.csv
df_jugadores=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2025/_Python/players_20.csv')
df_jugadores.set_index('short_name', inplace=True)
#Traemos las columnas que vamos
df_jugadores=df_jugadores[['player_url', 'long_name',	'age', 'dob',	'height_cm',	'weight_kg', 'nationality',	'club']]
df_jugadores
```

```python
#Con el metodo apply voy a calcular la raiz cuadrada de la columna edad "age"
df_jugadores['age'].apply(np.sqrt)
```

```python
#Con el metodo apply voy a calcular la raiz cuadrada de la columna edad "age"
df_jugadores['age'].apply(np.square)
```

```python
#Vamos a escribior euna función para calcular el IMC peso / estatura al cuadrado. Algo asi como kg /m2
def IMC(row):
    return row['weight_kg'] / ((row['height_cm'] / 100) ** 2)
```

```python
#Ahora vamos a llamr la función
df_jugadores.apply(IMC, axis=1)
```
