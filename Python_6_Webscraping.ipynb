```python
#la siguiente es la esctructura que podmeos ver en el link
"https://www.football-data.co.uk/mmz4281/"+"2425"+"/"+"E0"+".CSV"
```

```python
#Ya sé que la raíz es la misma entonces la guardo en una variable que se llame ruta
ruta='https://www.football-data.co.uk/mmz4281/'
```

```python
#Ahora lo que voy a hacer es crearme una lista con los campeonatos o termiaciones
campeonatos=['E0', 'E1', 'E2', 'E3', 'EC']
enlaces=[]

#Para que esto funcione debo recorrer la lista con un loop, a esta practica se le conoce como
#link dinamico, una vez recorra la lista debo reemplazar el valor para que el enlace funcione

for campeonato in campeonatos:
  df_links=pd.read_csv(ruta + '2425' + '/' + campeonato + '.csv')
  enlaces.append(df_links)
```
