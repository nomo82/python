**Parte 2: Manipulación de Datos**

```python
#Seleccionar columnas: Extraer solo las columnas rank, country_full y total_points.
df_fifa[['rank', 'country_full', 'total_points']]
```

```python
#Filtrar datos: Obtener las filas donde el rank sea menor a 10 (Top 10 equipos).
df_fifa[df_fifa['rank']<= 10][['country_full', 'rank']]
```

```python
#Filtrar con múltiples condiciones: Seleccionar equipos con más de 1500 puntos y que estén en el Top 20.
df_fifa[(df_fifa['total_points']>1500) & (df_fifa['rank']<=20)]
```

```python
#gregar una nueva columna: Crear una columna llamada "Rank Category" con estos valores:
#○ "Top" si el rank es menor a 10.
#○ "Medium" si el rank está entre 10 y 30.
#○ "Low" si el rank es mayor a 30.
condiciones_1=[
    df_fifa['rank']<10,
    (df_fifa['rank']>=10) & (df_fifa['rank']<=30),
    df_fifa['rank']> 30
]

valores_1=['Top', 'Medium', 'Low']
df_fifa['Rank Category']= np.select(condiciones_1, valores_1)
df_fifa
```

**Parte 5: Ejercicio Final - Análisis Avanzado**

```python
#Crear una nueva columna "Performance" basada en los total_points:
#○ "Excellent" si tiene más de 1600 puntos.
#○ "Good" si tiene entre 1400 y 1600 puntos.
#○ "Average" si tiene menos de 1400 puntos.
condiciones_2=[
    df_fifa['total_points']>1600,
    (df_fifa['total_points']>=1400) & (df_fifa['total_points']<=1600),
    df_fifa['total_points']< 1400
]

valores_2=['Excellent', 'Good', 'Average']
df_fifa['Performance']= np.select(condiciones_2, valores_2)
df_fifa
```

```python
#Contar cuántos equipos están en cada categoría de "Performance".
df_fifa['Performance'].value_counts()
```

```python
#Filtrar los equipos con "Excellent" performance y graficar sus puntos en un gráfico de barras (matplotlib o .plot(kind="bar")).

!pip install matplotlib   #Importación de librerias
import matplotlib as plt
```

```python
df_fifa_excellent= df_fifa[df_fifa ['Performance'] == 'Excellent']
df_fifa_excellent
```

```python
from matplotlib import pyplot as plt
df_fifa_excellent['total_points'].plot(kind='hist', bins=20, title='total_points')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
df_fifa_excellent.groupby('country_full').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
