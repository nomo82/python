```python
#Traemos el archivo DF Jugadores
#Leer el archivo CSV palyers 20.csv
df_jugadores=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2025/_Python/players_20.csv')
df_jugadores.head(10)
```

```python
#
df_jugadores=df_jugadores.set_index['short_name']
```

```python
#Vamos a traer las columnas que necesito y voy actualizar el DF
df_jugadores=df_jugadores[['player_url', 'long_name',	'age', 'dob',	'height_cm',	'weight_kg', 'nationality',	'club']]
df_jugadores
```

```python
#Vamos a traer al gato de Messi con el peso
df_jugadores.loc['L. Messi', 'weight_kg']
```

```python
#Vamos a modioficar el peso
df_jugadores.loc['L. Messi', 'weight_kg']=89 #Al colocar = 89 modifica el peso
```

```python
#Vanos a traer los datos de Messi con los dartos modificados
df_jugadores.loc['L. Messi', :]
```

```python
#Vamos a modioficar la altura de R7
df_jugadores.loc['Cristiano Ronaldo', 'height_cm']=190
```

```python
#Vamos a llamar los datos de R7
df_jugadores.loc['Cristiano Ronaldo', :]
```

```python
#Vamos a ver como actualizar el valor de toda una columna
df_jugadores.loc[:, 'height_cm']
```

```python
#Modificamos la altura por todo un valor de 190
df_jugadores.loc[:, 'height_cm']=190
df_jugadores
```

```python
#Como modificar el ultimo registro con el metodo .iloc
df_jugadores.iloc[1, :]
```

```python
#De esta manera puedo poner datos nulos en el DF
df_jugadores.iloc[-1, :]=np.nan
```

```python
#Llamamos el dato del DF
df_jugadores.iloc[-1, :]
```

```python
#Ahora vamoos a actualizar multiples valores por ejemplo actualizaremos a Messi y R7 al mismo tiempo estatura
df_jugadores.iloc[[0, 1], 4]=100
```

```python
#Vamos actualizar valores cuando se cumplen condiciones
#Primerto definimos las columnas a modificar con variables
columnas=['age', 'height_cm', 'weight_kg', 'nacionality', 'club']
df_jugadores.loc[df_jugadores['height_cm']>180, columnas]=0
```
