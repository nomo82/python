# Operaciones en Dataframes

```python
# crear una columna llamada "'Critic Rating'" usando la siguiente formula: Average Rating = (User Rating + Critic Rating)/2
df_bestseller['User Rating'] = pd.to_numeric(df_bestseller['User Rating'], errors='coerce')
df_bestseller['Critic Rating'] = pd.to_numeric(df_bestseller['Critic Rating'], errors='coerce')
df_bestseller['Average Rating'] = (df_bestseller['User Rating'] + df_bestseller['Critic Rating']) / 2
```

```python
#Imprimir los primeros 10 datos
df_bestseller.head(10)
```

```python
# usar la funcion "round" para redondear los valores del dataframe a 1 decimal y actualizar el dataframe
df_bestseller = df_bestseller.round(1)
df_bestseller.head()
```

```python
#Calcular 'Average Rating' ignorando NaN
df_bestseller['Average Rating'] = df_bestseller[['User Rating', 'Critic Rating']].mean(axis=1)
#Imprimir de nuevo BestSellers los primeros 20 valores
df_bestseller.head(7)
```

# Contar Valores

```python
# contar elementos en columna "Genre" por categoria y devolver la frecuencia relativa
df_bestseller['Genre'].value_counts()
```

```python
#Ahora creamos variable para 'Genre' y calculamos la frecuencia relativa
genre_counts = df_bestseller['Genre'].value_counts()
#Ahora creamos la variable para la frecuencia relatia
frecuence_relative = (genre_counts / genre_counts.sum()).round(2) #Round al final colocamos 2 para generar los decimales
print(frecuence_relative)
```
