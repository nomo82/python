```python
#Vamos a relaizar una tabla pivot sumando los registros. Observaremos en compra mas por genero
#Vamos a usar dentro del metodo 'pivot_table' un argumento que es rarto y se llama aggfunc = que me ayuda
#Vamos a contar
df_supermarket.pivot_table(index='Gender', aggfunc='count')
```

```python
#Ahora lo que vamos a ralizar es sumar en cada columna ppor gÃ©nero. Sin embargo tener presente que solo puedo sumar nuemros
df_supermarket.pivot_table(
    index='Gender',  # Agrupar por gÃ©nero
    values=['Quantity', 'Rating', 'Tax 5%', 'Total', 'Unit price', 'cogs', 'gross income', 'gross margin percentage'],  # Columnas a agregar
    aggfunc='sum'    # FunciÃ³n de agregaciÃ³n (suma)
)
```
