### 5. Manipulación de columnas (10 puntos)

```python
import pandas as pd

# Crear el DataFrame
df = pd.DataFrame({
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Precio': [100, 200, 150, 300, 250],
    'Cantidad': [5, 3, 8, 2, 4]
})

# a) Crear una nueva columna llamada 'Total' que sea el resultado de Precio * Cantidad
df['Total'] = df['Precio'] * df['Cantidad']

# b) Cambiar el nombre de la columna 'Producto' a 'Artículo'
df.rename(columns={'Producto': 'Artículo'}, inplace=True)

# Mostrar el DataFrame resultante
print(df)

```

### 6. Filtrado de datos (10 puntos)


```python
# Usando el DataFrame anterior:


# a) Filtrar y mostrar las filas donde el Precio sea mayor a 200
filtro_precio_mayor_200 = df[df['Precio'] > 200]
print("Filas con Precio mayor a 200:")
print(filtro_precio_mayor_200)

# b) Filtrar y mostrar las filas donde la Cantidad sea mayor a 3 Y el Precio menor a 200
filtro_cantidad_precio = df[(df['Cantidad'] > 3) & (df['Precio'] < 200)]
print("\nFilas con Cantidad mayor a 3 y Precio menor a 200:")
print(filtro_cantidad_precio)
```

### 7. Agrupación y agregación (10 puntos)

```python
# Dado el siguiente DataFrame:
import pandas as pd

ventas = pd.DataFrame({
    'Vendedor': ['Juan', 'María', 'Juan', 'Pedro', 'María', 'Pedro', 'Juan'],
    'Producto': ['A', 'B', 'C', 'A', 'A', 'B', 'B'],
    'Ventas': [100, 150, 200, 120, 300, 250, 180]
})

# a) Agrupar por Vendedor y calcular el total de ventas para cada uno
total_ventas_vendedor = ventas.groupby('Vendedor')['Ventas'].sum()
print("Total de ventas por Vendedor:")
print(total_ventas_vendedor)

# b) Agrupar por Vendedor y Producto, y calcular la media de ventas
media_ventas_vendedor_producto = ventas.groupby(['Vendedor', 'Producto'])['Ventas'].mean()
print("\nMedia de ventas por Vendedor y Producto:")
print(media_ventas_vendedor_producto)
```

### 9. Limpieza de datos (15 puntos)

```python
# Dado el siguiente DataFrame con valores faltantes:
import pandas as pd
import numpy as np

datos = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, 5],
    'D': [1, np.nan, np.nan, 4, 5]
})

# a) Cuenta cuántos valores faltantes hay en cada columna
datos.isnull().sum()

# b) Rellena los valores faltantes de la columna 'A' con la media de esa columna
media_A = datos['A'].mean()
datos['A'].fillna(media_A, inplace=True)

# c) Elimina las filas que tengan al menos 2 valores faltantes
datos_eliminados = datos.dropna(thresh=datos.shape[1] - 1)
datos_eliminados
```

### 10. Análisis final (15 puntos)

```python
# Supón que tienes un DataFrame llamado 'ventas_mensuales' con las columnas:
# 'Mes', 'Categoría', 'Ventas', 'Gastos'
# Escribe código para:

import pandas as pd

# Creación del ejemplo de un DF para el ejercicio
ventas_mensuales = pd.DataFrame({
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo'],
    'Categoría': ['Electrónica', 'Ropa', 'Electrónica', 'Ropa', 'Electrónica'],
    'Ventas': [1000, 800, 1200, 900, 1100],
    'Gastos': [600, 500, 700, 550, 650]
})
ventas_mensuales

```

```python
# a) Calcular el beneficio (Ventas - Gastos) para cada mes
ventas_mensuales['Beneficios'] = ventas_mensuales['Ventas'] - ventas_mensuales['Gastos']
ventas_mensuales
```

```python
# b) Encontrar la categoría con mayor beneficio total
mayor_beneficio = ventas_mensuales.groupby('Categoría')['Beneficio'].sum().idxmax()
mayor_beneficio
```

```python
# c) Crear una nueva columna 'Rentable' que sea True si el beneficio es positivo y False en caso contrario
ventas_mensuales['Rentable'] = ventas_mensuales['Beneficio'] > 0
ventas_mensuales
```
