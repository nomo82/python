```python
# Exportar libreria de datos de la carpeta impresos Cali
# Instalar bibliotecas necesarias
!pip install gspread oauth2client pandas pydrive2 --quiet

# Importar bibliotecas
from oauth2client.service_account import ServiceAccountCredentials
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
import pandas as pd
import gspread

# Autenticación de Google Colab
from google.colab import auth
auth.authenticate_user()

# Subir archivo JSON de credenciales
from google.colab import files
uploaded = files.upload()

# Configuración de acceso
scope = [
    'https://www.googleapis.com/auth/drive',
    'https://www.googleapis.com/auth/spreadsheets'
]
credentials = ServiceAccountCredentials.from_json_keyfile_name(list(uploaded.keys())[0], scope)
gc = gspread.authorize(credentials)

# Autenticación de PyDrive
gauth = GoogleAuth()
gauth.credentials = credentials
drive = GoogleDrive(gauth)

# ID de la carpeta de Google Drive
folder_id = '12LsURyeEMo1j0sU8_V254MNSz3zAufnD' #URL Carpeta Impresos Cali

# Obtener archivos de la carpeta
file_list = drive.ListFile({'q': f"'{folder_id}' in parents and mimeType='application/vnd.google-apps.spreadsheet'"}).GetList()

# Crear un DataFrame para almacenar los datos combinados
all_data = pd.DataFrame()

# Procesar cada archivo
for file in file_list:
    print(f"Procesando archivo: {file['title']}")
    try:
        # Abrir la hoja de Google Sheets
        sheet = gc.open_by_key(file['id']).sheet1

        # Obtener los encabezados y todos los datos
        headers = sheet.row_values(1)  # Encabezados
        data = sheet.get_all_values()  # Todos los datos

        # Crear un DataFrame con los datos
        df = pd.DataFrame(data[1:], columns=headers)  # Excluir la fila de encabezados

        # Agregar metadatos del archivo
        df['Nombre de archivo'] = file['title']

        # Concatenar los datos al DataFrame principal
        all_data = pd.concat([all_data, df], ignore_index=True)

    except Exception as e:
        print(f"Error procesando {file['title']}: {e}")
        continue

# Verificar y exportar los datos combinados con codificación 'utf-8'
if not all_data.empty:
    all_data.to_csv('datos_combinado_cali_impresos.csv', index=False, sep=';', encoding='utf-8')
    print('Datos exportados a datos_combinados_cali0_impresos.csv con separador ";" y codificación UTF-8')
else:
    print("No se encontraron datos para exportar.")
```

```python
# Instalar bibliotecas necesarias
!pip install gspread oauth2client pandas pydrive2 --quiet

# Importar bibliotecas
from oauth2client.service_account import ServiceAccountCredentials
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
import pandas as pd
import gspread
import time

# Autenticación de Google Colab
from google.colab import auth
auth.authenticate_user()

# Subir archivo JSON de credenciales
from google.colab import files
uploaded = files.upload()

# Configuración de acceso
scope = [
    'https://www.googleapis.com/auth/drive',
    'https://www.googleapis.com/auth/spreadsheets'
]
credentials = ServiceAccountCredentials.from_json_keyfile_name(list(uploaded.keys())[0], scope)
gc = gspread.authorize(credentials)

# Autenticación de PyDrive
gauth = GoogleAuth()
gauth.credentials = credentials
drive = GoogleDrive(gauth)

# ID de la carpeta de Google Drive
folder_id = '1AwqiXe8MUNG3_ZsjJ5CDm2nYg5VEz6_L'  # URL Carpeta Impresos Bogotá

# Obtener archivos de la carpeta
file_list = drive.ListFile({'q': f"'{folder_id}' in parents and mimeType='application/vnd.google-apps.spreadsheet'"}).GetList()

# Crear un DataFrame para almacenar los datos combinados
all_data = pd.DataFrame()

# Limitar las solicitudes por minuto (máximo 60)
requests_count = 0
REQUESTS_LIMIT = 60

# Procesar cada archivo
for file in file_list:
    print(f"Procesando archivo: {file['title']}")
    try:
        # Control de solicitudes
        if requests_count >= REQUESTS_LIMIT:
            print("Límite de solicitudes alcanzado. Esperando 60 segundos...")
            time.sleep(60)
            requests_count = 0

        # Abrir la hoja de Google Sheets
        sheet = gc.open_by_key(file['id']).sheet1
        requests_count += 1  # Contar solicitud

        # Obtener los encabezados y todos los datos
        data = sheet.get_all_values()
        headers = data[0]  # Primera fila como encabezados
        rows = data[1:]    # Excluir encabezados

        # Crear un DataFrame con los datos
        df = pd.DataFrame(rows, columns=headers)

        # Agregar metadatos del archivo
        df['Nombre de archivo'] = file['title']

        # Concatenar los datos al DataFrame principal
        all_data = pd.concat([all_data, df], ignore_index=True)

    except Exception as e:
        print(f"Error procesando {file['title']}: {e}")
        continue

# Verificar y exportar los datos combinados con codificación 'utf-8'
if not all_data.empty:
    output_file = 'datos_combinados_bogota_impresos.csv'
    all_data.to_csv(output_file, index=False, sep=';', encoding='utf-8')
    print(f'Datos exportados a {output_file} con separador ";" y codificación UTF-8')
else:
    print("No se encontraron datos para exportar.")
```

```python
# Exportar libreria de datos de la carpeta impresos Bogotá Año 2024
# Instalar bibliotecas necesarias
!pip install gspread oauth2client pandas pydrive2 --quiet

# Importar bibliotecas
from oauth2client.service_account import ServiceAccountCredentials
from pydrive2.auth import GoogleAuth
from pydrive2.drive import GoogleDrive
import pandas as pd
import gspread
from google.colab import auth
from google.colab import files
import datetime

# Autenticación de Google Colab
auth.authenticate_user()

# Subir archivo JSON de credenciales
uploaded = files.upload()

# Configuración de acceso
scope = [
    'https://www.googleapis.com/auth/drive',
    'https://www.googleapis.com/auth/spreadsheets'
]
credentials = ServiceAccountCredentials.from_json_keyfile_name(list(uploaded.keys())[0], scope)
gc = gspread.authorize(credentials)

# Autenticación de PyDrive
gauth = GoogleAuth()
gauth.credentials = credentials
drive = GoogleDrive(gauth)

# ID de la carpeta de Google Drive
folder_id = '1AwqiXe8MUNG3_ZsjJ5CDm2nYg5VEz6_L'  # URL Carpeta Impresos Bogotá

# Obtener archivos de la carpeta, filtrando por año 2024
file_list = drive.ListFile({'q': f"'{folder_id}' in parents and mimeType='application/vnd.google-apps.spreadsheet'"}).GetList()

# Crear un DataFrame para almacenar los datos combinados
all_data = pd.DataFrame()

# Procesar cada archivo
for file in file_list:
    try:
        # Obtener fecha de modificación
        modified_date = datetime.datetime.strptime(file['modifiedDate'], '%Y-%m-%dT%H:%M:%S.%fZ')

        # Filtrar solo archivos del año 2024
        if modified_date.year == 2024:
            print(f"Procesando archivo: {file['title']} (Modificado el {modified_date})")

            # Abrir la hoja de Google Sheets
            sheet = gc.open_by_key(file['id']).sheet1

            # Obtener los encabezados y todos los datos
            data = sheet.get_all_values()  # Todos los datos
            if data and len(data) > 1:
                headers = data[0]  # Encabezados
                # Validar encabezados únicos
                if len(headers) != len(set(headers)):
                    print(f"Advertencia: El archivo '{file['title']}' contiene encabezados duplicados. Ajustando nombres.")
                    headers = [f"{col}_{i}" if headers.count(col) > 1 else col for i, col in enumerate(headers)]

                # Crear un DataFrame con los datos
                df = pd.DataFrame(data[1:], columns=headers)  # Excluir la fila de encabezados

                # Agregar metadatos del archivo
                df['Nombre de archivo'] = file['title']

                # Concatenar los datos al DataFrame principal
                all_data = pd.concat([all_data, df], ignore_index=True)
            else:
                print(f"El archivo {file['title']} no contiene datos válidos.")
        else:
            print(f"Saltando archivo: {file['title']} (No es del año 2024)")

    except Exception as e:
        print(f"Error procesando {file['title']}: {e}")
        continue

# Verificar y exportar los datos combinados con codificación 'utf-8'
if not all_data.empty:
    all_data.to_csv('datos_combinados_bogota_impresos.csv', index=False, sep=';', encoding='utf-8')
    print('Datos exportados a datos_combinados_bogota_impresos.csv con separador ";" y codificación UTF-8')
else:
    print("No se encontraron datos para exportar del año 2024.")
```
