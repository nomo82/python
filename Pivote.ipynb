El método pivot en Python es una función útil en la librería pandas que se utiliza para reorganizar y transformar datos en un DataFrame. Este método es especialmente útil cuando se trabaja con datos tabulares y se desea cambiar la estructura de los datos, convirtiendo valores de filas en columnas o viceversa.

¿Qué hace el método pivot?
El método pivot toma tres parámetros principales:

index: La columna cuyos valores se convertirán en el índice del nuevo DataFrame.

columns: La columna cuyos valores se convertirán en las nuevas columnas del DataFrame.

values: La columna cuyos valores se utilizarán para rellenar el nuevo DataFrame.

Las tablas pivot (o tablas dinámicas) son una herramienta poderosa en el análisis de datos que permite resumir, reorganizar y agrupar grandes conjuntos de datos de manera eficiente. Son ampliamente utilizadas en herramientas como Excel, Google Sheets y librerías de programación como pandas en Python.

¿Qué son las tablas pivot?
Una tabla pivot toma datos de una tabla o DataFrame y los reorganiza en una nueva estructura, permitiendo:

Resumir datos: Agrupar y agregar datos (por ejemplo, sumar, promediar, contar).

Reorganizar datos: Cambiar la disposición de filas y columnas para obtener una vista más clara de la información.

Filtrar datos: Seleccionar subconjuntos de datos para enfocarse en información específica.

Usos comunes de las tablas pivot:
Calcular totales, promedios o recuentos de datos.

Analizar tendencias o patrones en los datos.

Comparar categorías o grupos.

Crear resúmenes para informes o visualizaciones.

Tablas pivot en Python (usando pandas)
En Python, la librería pandas proporciona la función pivot_table para crear tablas pivot. Esta función es muy flexible y permite personalizar cómo se agrupan y resumen los datos.
