```python
#Filtramos por Apple
df_computadores['Company']=='Apple'
df_computadores
```

```python
#Si deseo filtrar pos las dos condiciones anteriormente ejecutadas
#Que sea mayor a 2000 y que sea Apple debo usar AND o O
#AND en pandas es '&' y en pandas O es '|'
(df_computadores['Company']=='Apple') & (df_computadores['Price_euros']>2000)
```

```python
df_computadores[(df_computadores['Company']=='Apple') & (df_computadores['Price_euros']>2000)]
```

```python
#Ahora filtrar que sea DELL y que sea menor a 1000
(df_computadores['Company']=='Dell') & (df_computadores['Price_euros']<1000)
```

```python
df_computadores[(df_computadores['Company']=='Dell') & (df_computadores['Price_euros']<1000)]
```

```python
#Ahora vamos a ver el ejemplo con O '|'
#Realizamo el ejemplo con Apple
df_computadores[(df_computadores['Company']=='Apple') | (df_computadores['Price_euros']>2000)]
```

```python
#Ahora lo vamos a filtrar por las compaÃ±ias que cumplen con estos criterios
df_computadores[(df_computadores['Company']=='Apple') | (df_computadores['Price_euros']>2000)].value_counts('Company')
```

```python
df_computadores[(df_computadores['Company']=='Dell') | (df_computadores['Price_euros']<1000)].value_counts('Company')
```

```python
#Vamlmos a filtrar por computadores que sean Apple o Asus. Que cuesten menos de 2000
df_computadores[
    ((df_computadores['Company'] == 'Apple') | (df_computadores['Company'] == 'Asus')) &
    (df_computadores['Price_euros'] < 2000)
]
```

```python
#Vamlmos a filtrar por computadores que Apple. Que cuesten menos de 2000 y que sean mas de 13"
df_computadores [
    ((df_computadores['Company']=='Apple') | (df_computadores['Price_euros']<2000)) &
    df_computadores['Inches']='13')
]
```
