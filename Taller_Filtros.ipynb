**Parte 2: Creación de columnas**

```python
#5.	Clasificación por puntaje: Crea una nueva columna llamada "Clasificación puntos" que categorice a los equipos según su "total_points":
#"Muy alto" si tienen más de 1500 puntos.
#"Alto" si tienen entre 1000 y 1500 puntos.
#"Moderado" si tienen entre 500 y 1000 puntos.
#"Bajo" si tienen menos de 500 puntos. ¿Cuántos equipos hay en cada categoría?
condiciones_2=[
    df_fifa['total_points']>1500,
    (df_fifa['total_points']>1000) & (df_fifa['total_points']<=1500),
    (df_fifa['total_points']>500) & (df_fifa['total_points']<= 1000),
    df_fifa['total_points']<= 500
]

valores_2=['Muy Alto', 'Alto', 'Moderado', 'Bajo']
df_fifa['Clasificación puntos']= np.select(condiciones_2, valores_2)
df_fifa.value_counts('Clasificación puntos')
```

```python
#Niveles de promedio actual: Agrega una columna llamada "Nivel promedio actual" basada en "cur_year_avg":
#"Sobresaliente" si es mayor a 100.
#"Aceptable" si está entre 50 y 100.
#"Bajo" si es menor a 50. Realiza un conteo de cada nivel.
condiciones_3=[
    df_fifa['cur_year_avg']>100,
    (df_fifa['cur_year_avg']> 50) & (df_fifa['cur_year_avg']<=100),
    df_fifa['cur_year_avg']<= 50
]

valores_3=['Sobresaliente', 'Aceptable', 'Bajo']
df_fifa['Nivel promedio actual']= np.select(condiciones_3, valores_3)
df_fifa.head(5)
```

```python
#9.	Puntaje histórico alto: Filtra los equipos que tienen un "last_year_avg" mayor a 100 y "two_year_ago_avg" mayor a 100. ¿Cuántos equipos cumplen esta condición?
num_equipos_condicion = len(df_fifa[(df_fifa['last_year_avg'] > 100) & (df_fifa['two_year_ago_avg'] > 100)])
print(f"Número de equipos que cumplen la condición: {num_equipos_condicion}")
```

```python
#12.	Filtrado combinado: Usando isin(), filtra los equipos que pertenecen a "UEFA" y cuyo "rank" sea menor a 15. Muestra el resultado.
df_fifa[(df_fifa['confederation'].isin(['UEFA'])) & df_fifa['rank'].isin(range(1, 15))]
```

**Parte 5: Ejercicios integradores**

```python
#13.	Análisis de top y cambios: Filtra los equipos que están en el top 50 y que han tenido un cambio de ranking positivo ("rank_change" > 0). Muestra el total de equipos que cumplen esta condición y realiza un conteo por confederación.
df_fifa[(df_fifa['rank']<=50) & (df_fifa['rank_change']>0)].value_counts('country_full')
```

```python
#14.	Comparación de puntajes por año: Crea una nueva columna llamada "Puntaje histórico alto" que tenga el valor "Sí" si el "last_year_avg" es mayor que el "cur_year_avg"; en caso contrario, "No". ¿Cuántos equipos tuvieron un puntaje histórico más alto?
df_fifa['Puntaje histórico alto'] = np.where(df_fifa['last_year_avg'] > df_fifa['cur_year_avg'], 'Sí', 'No')
num_equipos_puntaje_alto = df_fifa['Puntaje histórico alto'].value_counts()['Sí']
print(f"Número de equipos con puntaje histórico alto: {num_equipos_puntaje_alto}")
```
