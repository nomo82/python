```python
#Siempre lo primero que debemos hacer es Importar pandas y numpy
import pandas as pd
import numpy as np
```

```python
#Ahora subimos el archivo desde Google Drive, ya no es necesario logearnos de nuevo
df_estudiantes=pd.read_csv('/content/drive/MyDrive/Alexander Niño.lnk/2024/_Phyton Curso/StudentsPerformance.csv')
df_estudiantes
```

```python
#Seguimos con la c reación de columnas usando el metdo asssign o el metodo insert
#Si quiero agregar varias columnas uso assign y tambien si quieres sobreescribir una columna
filosofia_score=np.random.randint(1, 100, size=1000)
fisica_score=np.random.randint(1, 100, size=1000)
```

```python
#Ahora pondre un indice en cada fila para saber en donde va el data frame
filosofia_nota=pd.Series(filosofia_score, index=np.arange(0,1000))
fisica_nota=pd.Series(fisica_score, index=np.arange(0,1000))
```

```python
#Ahora lo que voy hacer es poner las nuevas columnas a mi dataframe usando assign
df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
```

```python
#Para guardas las nuevas variables o columnas creadas debo llamar al dataframe
df_estudiantes=df_estudiantes.assign(filosofia_score=filosofia_nota, fisica_score=fisica_nota)
df_estudiantes
```

```python
#Ahora vamos el metodo insert, el metodo insert es diferente porque me permite esopecificar donde voy a colocar esa columna.
#Al usar el metodo no me devuelve copias debemos tener cuidado porque no me devuelve cambios
df_estudiantes.insert(1, 'ED_SCORE', fisica_nota)
```
